# Outlier DEtection

# Boxplot - Price Outliers

```python
plt.boxplot(df['Price_USD'],patch_artist=True)
plt.title('Price Outlier Detection')
plt.show()
```

Observation:-

# Box Plot- Revenue Outliers

```python
plt.boxplot(df['Revenue_USD'],patch_artist=True)
plt.title('Outlier Detection in the Revenue')
plt.show()
```

Observation:-

1= Some transactions generated unusually high revenue

2- These may represent bulk purchase or high price product

# IQR method for outlier detection

```python
q1=df['Revenue_USD'].quantile(0.25)
q3=df['Revenue_USD'].quantile(0.75)
IQR=q3-q1 #it represents the 50 percentile/ q2
lower_bound=q1-1.5*IQR
upper_bound=q3+1.5*IQR
outliers=df[(df['Revenue_USD']<lower_bound) | (df['Revenue_USD']>upper_bound)]
outliers
```

Observation: -

1- Outliers were identified using IQR method

2- high Revenue Values were detected outside noremal range
